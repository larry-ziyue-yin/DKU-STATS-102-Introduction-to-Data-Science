Load `WorldCups.csv`, and return which country hosted the FIFA would cup with the highest attendance?

# Problem 2 (0.5 point) 

Load `WorldCupMatches.csv`, and return the ratio of home team win.

You need to drop duplicate lines.

```python
import numpy as np
import pandas as pd

def myfunction():
    # YOUR CODE HERE
    df=pd.read_csv('WorldCupMatches.csv').dropna()
    length=len(df.index)
    win=0
    for i in range(len(df.index)):
        if df.iloc[i]['Home Team Goals']>df.iloc[i]['Away Team Goals']:
            win=win+1
        else:
            continue
    return win/length
```

# Problem 4 (0.5 point) 

Read `TerritoryReadMe.pdf` first to learn about the data in `PR.txt` and `TR.txt`.

In Puerto Rico and all other territories, find there are how many different names.

```python
import numpy as np
import pandas as pd

def cnt_names():
    # YOUR CODE HERE
    df1=pd.read_table('PR.txt',sep=',',header=None)
    df1.columns=['a','b','c','d','e']
    df2=pd.read_table('TR.txt',sep=',',header=None)
    df2.columns=['a','b','c','d','e']
    return len((df1['d'].value_counts())&(df2['d'].value_counts()))
```

```python
assert cnt_names() == 3179
```

# Problem 6 (0.5 point) 

Given the data of Chipotle Restaurant in `chipotle.tsv`, find the most-ordered item.

```python
import numpy as np
import pandas as pd

def find_item():
    # YOUR CODE HERE
    df=pd.read_csv('chipotle.tsv',delimiter = '\t')
    return df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(1).index.tolist()[0]
```
