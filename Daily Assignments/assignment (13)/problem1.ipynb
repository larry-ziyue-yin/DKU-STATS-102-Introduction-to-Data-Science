load WorldCupMatches.csv and WorldCupPlayers.csv to calculate all the matches that have goal in the first minute, and output the following dataframe.

![](ex5.png)

```python
import numpy as np
import pandas as pd
def myfunction():
    # YOUR CODE HERE
    df1=pd.read_csv('WorldCupMatches.csv').set_index('MatchID').dropna(how='all').drop_duplicates()
    df2=pd.read_csv('WorldCupPlayers.csv').set_index('MatchID').dropna(how='all').drop_duplicates()

    s=df2['Event'].dropna()
    a=[]
    alist=[]
    for i in range(len(s)):
        if "G1'" in s.iloc[i]:
            a.append(i)
    for m in a:
        alist.append(s.index[m])
    df=df2.loc[alist][['Player Name','Event']].dropna(how='any')
    num=[]
    for m in range(len(df)):
        if df['Event'].values[m].split('\'')[0]=='G1':
            num.append(m)
    DF=df.iloc[num]
    chart=pd.merge(df1,DF,on='MatchID')
    chart=chart[['Year','Datetime','Home Team Name','Home Team Goals','Away Team Goals','Away Team Name','Player Name']]
    return chart
```

```python
myfunction()
```
