# Problem 2

Draw the Maurer rose curve (https://en.wikipedia.org/wiki/Maurer_rose) when $n=6, d=71$.

Hint: $\frac{\pi}{180}\approx0.0174533$, polar coordinate system to Cartesian coordinate system (https://en.wikipedia.org/wiki/Polar_coordinate_system)

![](assignment14.png)

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

def myplot():
    n = 6
    d = 71
    # YOUR CODE HERE
    rad=0.0174533
    for i in range(361):
        x=[np.cos(d*i*rad)*np.sin(d*n*i*rad),np.cos((i+1)*d*rad)*np.sin((i+1)*d*n*rad)]
        y=[np.sin(d*i*rad)*np.sin(d*n*i*rad),np.sin((i+1)*d*rad)*np.sin((i+1)*d*n*rad)]
        plt.title('Maurer rose')
        plt.plot(x,y,color='b')
        plt.axis('equal')
```

```python
myplot()
#subject to manually checking
```
