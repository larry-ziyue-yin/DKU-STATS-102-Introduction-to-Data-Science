```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
import pandas as pd
import numpy as np
label = {1:'Clear', 2:'Few clouds', 3:'Broken clouds', 4:'Cloudy', 7:'Rain', 10:'Rain with thunderstorm', 26:'Snowfall'}
marker = ['o', 'x', '+', 'v', '>', 's', 'd']
# YOUR CODE HERE
df=pd.read_csv('london_merged.csv')
df=df.groupby(by=['weather_code']).sum().sort_values(by['cnt'],ascending=False)#cnt是count的意思。
y=df['cnt']
x=df.index.astype(int)
for i in range(y.size):
    plt.scatter(x[i],y[i],marker=marker[i],label=label[x[i]])
```
